```python
import pandas as pd
import numpy as np
from functools import reduce
df=pd.read_csv('/content/Databases/BL-Flickr-Images-Book.csv')
df.head()
```

```python
unwanted_charecters = ['[',',','-']
def clean_dates(item):
  dop = str(item.loc['Date of Publication'])
  if dop == 'nan' or dop[0] == '[':
    return np.NaN
  for charecter in unwanted_charecters:
    if charecter in dop:
      charecter_index = dop.find(charecter)
      dop = dop[:charecter_index]
    return dop
df['Date of Publication']=df.apply(clean_dates,axis=1)
```

```python
def clean_dates1(dop):
  dop = str(dop)
  if dop.startswith('[') or dop == 'nan':
    return 'NaN'
  for charecter in unwanted_charecters:
    if charecter in dop:
      charecter_index = dop.find(charecter)
      dop = dop[:charecter_index]
  return dop
df['Date of Publication']=df['Date of Publication'].apply(clean_dates1)
df.head()
```

```python
def clean_titles(title):
  if title =='nan':
    return 'NaN'
  if title[0] == '[':
    title=title[1: title.find(']')]
    if 'by' in title:
      title=title[1: title.find('by')]
    elif 'By' in title:
      title=title[1: title.find('By')]
    if title[0] == '[':
      title=title[1: title.find('[')]
    title=title[:-2]
    title=list(map(str.capitalize,title.split()))
    return ' '.join(title)
df['Title']=df['Title'].apply(clean_titles)
df.head()
```
