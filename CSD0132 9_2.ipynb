```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
print("step 1")
```

```python
fig = plt.figure()
ax=plt.axes()
ax.grid()
```

```python
fig = plt.figure()
ax=plt.axes()
x=np.linspace(0,9,100)
print(x)
ax.plot(x, np.sin(x))
```

```python
fig = plt.figure()
ax=plt.axes()
x=np.linspace(0,9,100)
print(x)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
```

```python
fig = plt.figure()
ax=plt.axes()
x=np.linspace(0,9,100)
print(x)
ax.plot(x, np.sin(x))
ax.plot(x, np.tan(x))
ax.set_title('Simple Plot')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
```

```python
fig = plt.figure()
ax=plt.axes()
x=np.linspace(0,9 ,50)

ax.plot(x, np.sin(x),label='sin',color='black',linestyle='dashed')
ax.plot(x, np.cos(x),label='cos',color='b',linestyle='dashdot')
ax.plot(x, np.tan(x),label='tan',color='g',linestyle='dotted')
ax.set_title('Multiple Lines')
ax.set_xlabel('Period')
ax.set_ylabel('Amplitude')
ax.legend()
plt.show()
```
