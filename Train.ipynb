```python
import pandas as pd
df = pd.read_csv('/content/dataset/train.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.tail(4)
```

```python
age_mean_value=df['Age'].mean()
df['Age']=df['Age'].fillna(age_mean_value)
df.tail(4)
```

```python
df.drop("Cabin",axis=1,inplace=True)
df.head()
```

```python
filtered_age=df[df.Age>50]
filtered_age
```

```python
sorted_passengers=df.sort_values('Age',ascending=True,kind='heapsort')
sorted_passengers
```

```python
merged_df=pd.merge(df.head(2),df.tail(2),how='outer',indicator=True)
merged_df
```

```python
group_df=df.groupby('Survived')
for key,item in group_df:
  print(group_df.get_group(key),"\n\n")
```
